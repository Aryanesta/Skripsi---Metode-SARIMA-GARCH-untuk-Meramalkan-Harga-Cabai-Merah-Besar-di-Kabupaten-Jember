```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from statsmodels.tsa.stattools import acf, pacf

df = pd.read_csv('/content/harga_cabe_pasar_1857.csv', index_col='tanggal', parse_dates=True)

max_lag = 929
data_series = df["harga"].dropna()

nilai_acf = acf(data_series, nlags=max_lag)
nilai_pacf = pacf(data_series, nlags=max_lag, method="ywadjusted")

conf_interval = 1.96 / np.sqrt(len(data_series))

lags = np.arange(max_lag + 1)

# Visualisasi ACF
fig = go.Figure()

fig.add_trace(go.Bar(x=lags, y=nilai_acf, name="ACF", marker_color="royalblue"))
fig.add_hline(
    y=conf_interval,
    line_dash="dash",
    line_color="red",
    annotation_text="Batas Atas",
)
fig.add_hline(
    y=-conf_interval,
    line_dash="dash",
    line_color="red",
    annotation_text="Batas Bawah",
)

fig.update_layout(
    title=f"Korelogram ACF (Hingga Lag {max_lag})",
    xaxis_title="Lag",
    yaxis_title="Koefisien Autokorelasi",
    barmode="overlay",
)

fig.show()

plt.figure(figsize=(18, 6))
plt.bar(lags, nilai_acf, color="darkblue", width=0.8, label="ACF")
plt.axhline(y=conf_interval, color="red", linestyle="--", label="Significance")
plt.axhline(y=-conf_interval, color="red", linestyle="--")

plt.title(f"Grafik ACF (0 - {max_lag})", fontsize=14)
plt.xlabel("Lag", fontsize=12)
plt.ylabel("Autokorelasi", fontsize=12)
plt.xlim(-10, max_lag + 10)
plt.grid(axis="x", alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

print("\n=== NILAI ACF ===")
for lg in range(0, max_lag + 1, 1):
    if lg <= max_lag:
        print(f"Lag {lg:4d}: {nilai_acf[lg]:.4f}")
```

```python
# Visualisasi PACF
fig_pacf = go.Figure()
fig_pacf.add_trace(go.Bar(
    x=lags,
    y=nilai_pacf,
    name="PACF",
    marker_color="teal",
    opacity=0.8
))

fig_pacf.add_hline(y=conf_interval, line_dash="dash", line_color="red", line_width=1.5, annotation_text="95% CI Upper")
fig_pacf.add_hline(y=-conf_interval, line_dash="dash", line_color="red", line_width=1.5, annotation_text="95% CI Lower")

fig_pacf.update_layout(
    title=f"Korelogram PACF (0 - {max_lag})",
    xaxis_title="Lag",
    yaxis_title="Koefisien Parsial Autokorelasi",
    xaxis=dict(showgrid=True, nticks=20),
    yaxis=dict(range=[-1.1, 1.1]),
    template="plotly_white",
    barmode="overlay",
)

fig_pacf.show()

plt.figure(figsize=(18, 6))
plt.bar(lags, nilai_pacf, color="darkcyan", width=0.7, label="PACF", alpha=0.9)
plt.axhline(y=conf_interval, color="red", linestyle="--", linewidth=1.5, label="Batas Signifikansi 95%")
plt.axhline(y=-conf_interval, color="red", linestyle="--", linewidth=1.5)

plt.title(f"Grafik PACF (Partial Autocorrelation) - Lag 0 hingga {max_lag}", fontsize=14)
plt.xlabel("Lag", fontsize=12)
plt.ylabel("Koefisien PACF", fontsize=12)
plt.xlim(-10, max_lag + 10)
plt.ylim(-1, 1)
plt.grid(axis="x", linestyle='-', alpha=0.2)
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()

print("\n=== NILAI PACF ===")
for lg in range(0, max_lag + 1, 1):
    print(f"Lag {lg:4d}: {nilai_pacf[lg]:.4f}")
```
